```python
import eikon as ek

ek.set_app_key('12345') # enter your refinitive api
```

```python
import pandas as pd
from datetime import datetime, timedelta

# List of top 20 companies (use RICs or identifiers supported by the API)
top_20_companies = ['BAC.N', 'SNAP.K','PFE.N','NIO.N'
    'NVDA.O', 'TSLA.O', 'SMCI.O', 'PLTR.O', 'INTC.O', 'AAPL.O', 
    'MARA.O',  'AMD.O', 'SOFI.O', 'PLUG.O','LCID.O', 'SOUN.O','AMZN.O', 'RIVN.O', ,'CLSK.O', 'TLRY.O'] 
    
# Date range
date_from = datetime.strptime("2024-01-01T05:00:00", "%Y-%m-%dT%H:%M:%S")
date_to = datetime.strptime("2024-12-31T17:00:00", "%Y-%m-%dT%H:%M:%S")

# Time interval for splitting (12 hours)
time_interval = timedelta(hours=12)

# Function to fetch news headlines
def fetch_news_headlines(company, start_time, end_time):
    try:
        headlines = ek.get_news_headlines(
            query=f'R:{company} IN ENGLISH',  # Query for the company
            date_from=start_time.strftime("%Y-%m-%dT%H:%M:%S"),
            date_to=end_time.strftime("%Y-%m-%dT%H:%M:%S"),
            count=100  # Maximum allowed per request
        )
        return headlines
    except Exception as e:
        print(f"Error fetching data for {company}: {e}")
        return pd.DataFrame() 

```

```python
# Collect news headlines for all companies
all_news = pd.DataFrame()

for company in top_20_companies:
    current_time = date_from
    while current_time < date_to:
        next_time = current_time + time_interval
        if next_time > date_to:
            next_time = date_to

        print(f"Fetching news for {company} from {current_time} to {next_time}")
        news = fetch_news_headlines(company, current_time, next_time)

        if not news.empty:
            news['company'] = company
            all_news = pd.concat([all_news, news], ignore_index=True)
        current_time = next_time



```

```python
# Save to CSV
all_news.to_csv("top_20_companies_news.csv", index=False)

print("Data collection complete. Total news collected:", len(all_news))
```
